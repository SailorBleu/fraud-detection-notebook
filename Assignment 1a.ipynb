```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df = pd.read_csv('Fraud.csv')

# Assuming 'df' is your DataFrame
pd.set_option('display.float_format', '{:.2f}'.format)

# Now display your DataFrame
print(df.describe())
```

```python
# Display basic information about the dataset
df.info()

# Calculate summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Count the number of fraudulent vs non-fraudulent transactions
class_distribution = df['isFraud'].value_counts()

# Plot the class distribution
sns.barplot(x=class_distribution.index, y=class_distribution.values)
plt.title('Distribution of Fraudulent vs Non-Fraudulent Transactions')
plt.xlabel('type')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Fraudulent', 'Fraudulent'])
plt.show()
```

```python
# Plot histograms for key variables
df['amount'].hist(bins=50, edgecolor='black')
plt.title('Distribution of Transaction Amount')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()

df['oldbalanceOrg'].hist(bins=50, edgecolor='black')
plt.title('Distribution of Original Old Balance')
plt.xlabel('Old Balance Original')
plt.ylabel('Frequency')
plt.show()

df['newbalanceOrig'].hist(bins=50, edgecolor='black')
plt.title('Distribution of Original New Balance')
plt.xlabel('New Balance Original')
plt.ylabel('Frequency')
plt.show()

df['oldbalanceDest'].hist(bins=50, edgecolor='black')
plt.title('Distribution of Destination Old Balance')
plt.xlabel('Old Balance Destination')
plt.ylabel('Frequency')
plt.show()

df['newbalanceDest'].hist(bins=50, edgecolor='black')
plt.title('Distribution of Destination New Balance')
plt.xlabel('New Balance Destination')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for Transaction Amount
sns.boxplot(x=df['amount'])
plt.title('Box Plot of Transaction Amount')
plt.show()

# Box plot for Original Old Balance
sns.boxplot(x=df['oldbalanceOrg'])
plt.title('Box Plot of Original Old Balance')
plt.show()

# Box plot for Original New Balance
sns.boxplot(x=df['newbalanceOrig'])
plt.title('Box Plot of Original New Balance')
plt.show()

# Box plot for Destination Old Balance
sns.boxplot(x=df['oldbalanceDest'])
plt.title('Box Plot of Destination Old Balance')
plt.show()

# Box plot for Destination New Balance
sns.boxplot(x=df['newbalanceDest'])
plt.title('Box Plot of Destination New Balance')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
